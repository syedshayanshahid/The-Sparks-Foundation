```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\PC\Desktop\TSF-INTERN\global_terrorism.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# Basic information about the dataset
df.info
```

```python
# Summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize the distribution of terrorist attacks over the years
plt.figure(figsize=(15, 6))
sns.countplot(x='iyear', data=df, palette='viridis')
plt.title('Number of Terrorist Attacks Over the Years')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the countries with the highest number of terrorist attacks
plt.figure(figsize=(15, 6))
sns.barplot(x=df['country_txt'].value_counts()[:15].index, y=df['country_txt'].value_counts()[:15].values, palette='viridis')
plt.title('Top 15 Countries with the Highest Number of Terrorist Attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the most common attack types
plt.figure(figsize=(15, 6))
sns.countplot(x='attacktype1_txt', data=df, palette='viridis')
plt.title('Most Common Attack Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the target types
plt.figure(figsize=(15, 6))
sns.countplot(x='targtype1_txt', data=df, palette='viridis')
plt.title('Most Common Target Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the weapon types used in terrorist attacks
plt.figure(figsize=(15, 6))
sns.countplot(x='weaptype1_txt', data=df, palette='viridis')
plt.title('Most Common Weapon Types Used in Attacks')
plt.xticks(rotation=45)
plt.show()
```
