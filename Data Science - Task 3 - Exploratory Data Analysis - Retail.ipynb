```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv(r'C:\Users\PC\Desktop\TSF-INTERN\sample_superstore.csv')

# Display basic information about the dataset
print(df.info())

# Display summary statistics
print(df.describe())
```

```python
# Total Sales and Profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

# Average Sales and Profit per order
average_sales_per_order = df['Sales'].mean()
average_profit_per_order = df['Profit'].mean()

# Visualize Sales and Profit distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Sales'], 
             bins=20, 
             kde=True, 
             color='blue', 
             label='Sales')
sns.histplot(df['Profit'], 
             bins=20, 
             kde=True, 
             color='green', 
             label='Profit')
plt.title('Distribution of Sales and Profit')
plt.legend()
plt.show()
```

```python
# Total Sales and Profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

# Average Sales and Profit per order
average_sales_per_order = df['Sales'].mean()
average_profit_per_order = df['Profit'].mean()

# Explore sales and profit by different categories
category_sales = df.groupby('Category')['Sales'].sum()
category_profit = df.groupby('Category')['Profit'].sum()

# Visualize category-wise sales and profit
plt.figure(figsize=(12, 6))
category_sales.plot(kind='bar', color='blue', label='Sales')
category_profit.plot(kind='bar', color='orange', label='Profit')
plt.title('Category-wise Sales and Profit')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
# Average discount
average_discount = df['Discount'].mean()

# Visualize the distribution of discounts
plt.figure(figsize=(12, 6))
sns.histplot(df['Discount'], bins=20, kde=True, color='orange', label='Discount')
plt.title('Distribution of Discounts')
plt.legend()
plt.show()
```

```python
# Analyze the impact of discount on profit
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount', y='Profit', data=df, color='red')
plt.title('Impact of Discount on Profit')
plt.show()
```

```python
# Analyze sales and profit by region
region_sales = df.groupby('Region')['Sales'].sum()
region_profit = df.groupby('Region')['Profit'].sum()

# Visualize region-wise sales and profit
plt.figure(figsize=(12, 6))
region_sales.plot(kind='bar', color='blue', label='Sales')
region_profit.plot(kind='bar', color='green', label='Profit')
plt.title('Region-wise Sales and Profit')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
# Analyze sales and profit by sub-category
sub_category_sales = df.groupby('Sub-Category')['Sales'].sum()
sub_category_profit = df.groupby('Sub-Category')['Profit'].sum()

# Visualize sub-category-wise sales and profit
plt.figure(figsize=(15, 6))
sub_category_sales.plot(kind='bar', color='orange', label='Sales')
sub_category_profit.plot(kind='bar', color='red', label='Profit')
plt.title('Sub-Category-wise Sales and Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
# Analyze sales and profit by ship mode
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum()
ship_mode_profit = df.groupby('Ship Mode')['Profit'].sum()

# Visualize ship mode-wise sales and profit
plt.figure(figsize=(12, 6))
ship_mode_sales.plot(kind='bar', color='yellow', label='Sales')
ship_mode_profit.plot(kind='bar', color='purple', label='Profit')
plt.title('Ship Mode-wise Sales and Profit')
plt.xlabel('Ship Mode')
plt.ylabel('Amount')
plt.legend()
plt.show()
```
